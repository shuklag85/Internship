```python
import requests
from bs4 import BeautifulSoup

def dinner(url):
    restaurant_data = []
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all restaurant containers
        restaurants = soup.find_all('div', class_='restnt-card restaurant')
        
        for restaurant in restaurants:
            
            name_element = restaurant.find('div', class_='restnt-info cursor')
            
            Cuisn_element = restaurant.find('span', class_='double-line-ellipsis')
            Cuisn_all = Cuisn_element.find_all('a')
            Cuisn=""
            
            Rating_element= restaurant.find('div', class_='img-wrap')
            Rating_all= Rating_element.find_all('div')
            Ratng=""
            
            Loc=name_element.find('div',class_="restnt-loc ellipsis")
            Loc_all=Loc.find_all('a')
            Locc=""
            
            image_url = restaurant.find('img', class_='no-img')
            
            if name_element and Cuisn_element and image_url :
                name = name_element.find('a').text.strip()
                image_url_all = image_url['data-src']
                
                for a in Cuisn_all:
                    Cuisn = Cuisn+a.text.strip()+" "
                for b in Loc_all:
                    Locc = Locc+b.text.strip()+" " 
                for c in Rating_all:
                    Ratng=Ratng+c.text.strip()+" "
                
            
                    

                restaurant_data.append({
                    'Restaurant Name': name,
                    'Cusine_name': Cuisn.strip(),
                    'Location': Locc.strip(),
                    'Ratings' : Ratng.strip(),
                    'Image URL': image_url_all
                    
                })
                    
            else:
                missing_elements = []
                if not name_element:
                    missing_elements.append('Name')
                
                
                print(f"Some elements are missing for a restaurant ({', '.join(missing_elements)}), skipping...")
    
    return restaurant_data

# Example URL of Dineout website
dineout_url = 'https://www.dineout.co.in/bangalore-restaurants'
restaurant_details = dinner(dineout_url)

# Print scraped details
for restaurant in restaurant_details:
    print(restaurant)
```
