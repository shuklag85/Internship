results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand  
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and  
‚ÄúProduct URL‚Äù. In case, if any of the details are missing for any of the product then replace it by ‚Äú-‚Äú.

```python
Brand_Name=[]
Prod_Name=[]
Price=[]
Return=[]
Delivery=[]
Availability=[]
URL=[]

Name=driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
Prz=driver.find_elements(By.XPATH,'//span[@class="a-price"]/span/span[2]')
Del=driver.find_elements(By.XPATH,'//span[@class="a-color-base a-text-bold"]')
ur=driver.find_elements(By.XPATH,'//img[@class="s-image"][1]')
print(len(Name))
print(len(Prz))
print(len(Del))
print(len(ur))
for i in range(len(Name)):
    N=Name[i].text
    p=Prz[i].text
    D=Del[i].text
    U=ur[i].get_attribute('src')
    
    for a in ur:
        driver.get(a)
        time.sleep(3)
        try:
            br=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[11]/div/div[3]/div/div/div[1]/div[1]/p/span')
            Brand_Name.append(br.text)
        except NoSuchElementException:
            Brand_Name.append('Not found')

            
    #print(N)
    #print(p)
    
    Prod_Name.append(N)
    Price.append(p)
    Delivery.append(D)
    URL.append(U)
    
df=pd.DataFrame({
    "Prod_Name":Prod_Name,
    'Price':Price,
    'Delivery':Delivery,
    'URL':URL,
    'Brand_Name':Brand_Name
})
df
```

Write a python program to access the search bar and search button on images.google.com and scrape 10 
images each for keywords ‚Äòfruits‚Äô, ‚Äòcars‚Äô and ‚ÄòMachine Learning‚Äô, ‚ÄòGuitar‚Äô, ‚ÄòCakes‚Äô.  

```python
def scrape_images_for_keyword(driver, keyword):
    # Navigate to Google Images
    driver.get("https://www.google.com/imghp")

    # Find the search bar element and enter the keyword
    search_box = driver.find_element(By.NAME, "q")
    search_box.clear()
    search_box.send_keys(keyword)
    search_box.submit()

    # Wait for images to load
    time.sleep(5)  # Increase waiting time if needed

    # Scroll down to load more images
    num_scrolls = 5
    for _ in range(num_scrolls):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)

    # Find all image elements on the page
    image_elements = driver.find_elements(By.XPATH, '//div[@id="islrg"]//img')

    # Extract image URLs from the elements
    image_urls = []
    for img in image_elements:
        src = img.get_attribute('src')
        if src and 'http' in src:  # Ensure valid image URLs
            image_urls.append(src)

    print(image_urls[:10]) # Return only the first 10 image URLs
```

Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted 
from any YouTube Video


```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up Selenium WebDriver (Chrome)
driver = webdriver.Chrome()
driver.get('https://www.youtube.com/watch?v=67gRtgGL4dg')
driver.maximize_window()
time.sleep(5)  # Allow time for the video page to load

# Scroll down to load comments
for _ in range(3):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(3)  # Wait for comments to load

# Wait for all comment elements to be present
comments = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '#content-text')))

# Extract comments text
data = [comment.text for comment in comments]

# Print or process the scraped comments
print("Scraped Comments:")
print(data)

# Quit the WebDriver session
driver.quit()
```
